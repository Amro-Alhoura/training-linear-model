# Linear regression (Normal equation)

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Creating data:

```python
np.random.seed(0)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1) # to mske noise on linear regression
```

```python
plt.plot(x, y, '.')
plt.xlabel('X', fontsize = 18)
plt.ylabel('Y', rotation = 0, fontsize = 18)
plt.axis([0, 2, 0, 15]);
```

### Normal equation code:

```python
def normal_equation(X, Y):
    X_bias = np.append(np.ones((100, 1)), X, axis = 1) # Adding X(0) bias col
    W = np.linalg.inv(X_bias.T.dot(X_bias)).dot(X_bias.T).dot(Y) # Normal equation
    return W

normal_equation(x, y)
```

### Prediction code:

```python
def prediction(new_X, W):
    new_X_b = np.append(np.ones((new_X.shape[0], 1)), new_X, axis = 1) # Adding X(0) bias col
    predictions = new_X_b.dot(W) # Linear equation: y = b + w * x
    return predictions

w = normal_equation(x, y)
new_x = np.array([[0], [1.3], [2]])
predictions = prediction(new_x, w)
predictions
```

```python
plt.plot(x ,y ,'.')
plt.plot(new_x, predictions, '-', label = 'Prediction')
plt.xlabel('X', fontsize = 18)
plt.ylabel('Y', rotation = 0, fontsize = 18)
plt.axis([0, 2, 0, 15]);
```

### Compare result with siket-learn linear regression:

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression() # LinearRegression use the SVD approach
lin_reg.fit(x, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
lin_reg.predict(new_x)
```
